```python
df.isnull().sum()
```

```python
df["tree.age"].fillna(df["tree.age"].mean(), inplace=True)
df["l.score"].fillna(df["l.score"].mean(), inplace=True)
df["rain"].fillna(df["rain"].mean(), inplace=True)
df["wind.intensity"].fillna(df["wind.intensity"].mean(), inplace=True)
df["humidity"].fillna(df["humidity"].mean(), inplace=True)
```

```python
df.isnull().sum()
```

We can check that there are missing values and we have treated them by replacing the mean or average of that particular feature,
We took the mean instead of dropping the rows as it is a small dataset.
