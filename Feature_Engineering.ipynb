Dropped 'collector.id' feature as it does not contribute significantly to prediction accuracy. 

```python
targetname ="fire"
allfeaturenames = df.columns[df.columns!=targetname]
x = df[allfeaturenames]
y= df[targetname]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

Split the data into train-test and allocating the features and target variables

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import numpy as np
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [9])], remainder='passthrough')

# Fit and transform X_train
x_train = np.array(ct.fit_transform(x_train))

# Transform X_test
x_test = np.array(ct.transform(x_test))
```

Transforming Catergorical Variable 'time.of.day' into Numerical variable

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train[:,9:]=sc.fit_transform(x_train[:,9:])
x_test[:,9:]=sc.transform(x_test[:,9:])
```

Using Standard Scaler to Standardize the dataset. 
Standardization transforms data to have a mean of 0 and a standard deviation of 1, ensuring features contribute equally to model performance.
