# Gradient Boost

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
param_grid = {
    'n_estimators': [50, 100, 200],  # Number of boosting stages (trees)
    'learning_rate': [0.01, 0.1, 0.5],  # Step size shrinkage used to prevent overfitting
    'max_depth': [3, 5, 7],  # Maximum depth of the individual trees
    'min_samples_split': [2, 5, 10],  # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4]  # Minimum number of samples required to be at a leaf node
}
gb_classifier = GradientBoostingClassifier()
grid_search = GridSearchCV(estimator=gb_classifier, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
```

```python
grid_search.fit(x_train, y_train)
print("Best Parameters:", grid_search.best_params_)
```

```python
print("Best Estimator:", grid_search.best_estimator_)
```

```python
y_pred = grid_search.best_estimator_.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy:", accuracy)
```

# K Nearest Classifier

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
param_grid = {
    'n_neighbors': range(1, 21),  # Experiment with different ranges
    'metric': ['euclidean', 'manhattan', 'minkowski'],  # Experiment with different distance metrics
    'weights': ['uniform', 'distance']  # Experiment with different weighting schemes (optional)
}
```

```python
knn_model = KNeighborsClassifier()
grid_search = GridSearchCV(estimator=knn_model, param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)
```

```python
print("Best parameters:", grid_search.best_params_)
print("Best training score:", grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_
predictions = best_model.predict(x_test)

# Evaluate the model's accuracy on the testing set
accuracy = accuracy_score(y_test, predictions)
print("Test set accuracy:", accuracy)
```
